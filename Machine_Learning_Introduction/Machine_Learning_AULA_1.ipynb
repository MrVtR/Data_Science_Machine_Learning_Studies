````python
#Visualizar % de dados faltantes

import pandas as pd
import numpy as np

df  =  pd.read_csv('/content/dirtydata.csv')

for col in df.columns:
    faltantes = np.sum(df[col].isnull())
    print('{} - {}%'.format(col, round((faltantes/len(df))*100)))
 
````

````python
#Exemplo - substitui NULL por um número apenas para colunas especificadas

df["Calories"].fillna(130, inplace = True) #número 130
print(df)
````

````python
#Substitui usando média, mediana ou moda
#mean() median()e mode()

x = df["Calories"].mean()
print(x)

df["Calories"].fillna(x, inplace = True)
print(df)
````

````python
#Na linha 26 a coluna 'Date' está no formato string:
df[26:]
````

````python
#Converte para formato datetime
df['Date'] = pd.to_datetime(df['Date'])
df['Date'].iloc[26:]
````

````python
#Remove as linhas com um valor NULL (NaT) na coluna "Data" - linha 22 
df.dropna(subset=['Date'], inplace = True) 
````

````python
#Dados incorretos

#Em nosso conjunto de dados, na linha 7, 'Duration' é 450 (poderia ser um outlier, erro de digitação)
#já que os outros valores está entre 30 e 60
#Podemos corrigir o valor diretamente na linha 7
df.loc[7, 'Duration'] = 45
df[7:]
````

````python
#Loop por todos os valores na coluna "Duration"
#Se o valor for superior a 120, defina-o como 120:

for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
````

````python
#Exclua as linhas em que a "Duration" seja superior a 120:
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
````

````python
#Removendo dados duplicados
#Observando nosso conjunto de dados de teste, podemos assumir que as linhas 11 e 12 são duplicatas

print(df.duplicated())

#Remova todas as duplicatas
df.drop_duplicates(inplace = True)
````

````python
#import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler

iris = datasets.load_iris()

#Dividindo em X e y (Class labels)
X = iris.data  
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size = 0.5)
````

````python
# técnica under-sampling
rus = RandomUnderSampler()
X_res, y_res = rus.fit_resample(X_train, y_train)

print(X_train.shape)
print(y_train.shape)

print(X_res.shape)
print(y_res.shape)
````

````python
#PCA - Principal Component Analisys

#import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D

iris = datasets.load_iris()

print(iris)
````

````python
#Dividindo em X e y (Class labels)

X = iris.data  
y = iris.target

print(y)
````

````python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)

print(X_std)
````

````python
from sklearn.decomposition import PCA as sklearnPCA

sklearn_pca = sklearnPCA(n_components=3)
Y_sklearn = sklearn_pca.fit_transform(X_std)

print(Y_sklearn)
````

````python
fig = plt.figure(1, figsize=(8, 6))

ax = Axes3D(fig, elev=-150, azim=110)

ax.scatter(
    Y_sklearn[:, 0],
    Y_sklearn[:, 1],
    Y_sklearn[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
ax.set_title("PCA 3 Componentes")
ax.set_xlabel("1 Componente")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2 Componente")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3 Componente")
ax.w_zaxis.set_ticklabels([])

plt.show()
````

Seleção de Atributos (features) 

```
O valor mostrado para cada correlação vai de:
-1 que indica uma correlação negativa (quando se movem em direções opostas) 
exemplo: quanto mais alguém estuda, maior é sua nota na prova

+1 uma correlação positiva (quando se movem na mesma direção)
# exemplo: quanto maior a velocidade de um carro, menor é o tempo para chegar ao destino.

0 quando a correlação é igual a zero, significa que não é possível determinar um senso de covariação
```

Download Íris [Dataset](https://drive.google.com/file/d/1zU372zcHbv3ezRIUYONAkZBz5tsgCnM8/view?usp=sharing)

````python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/iris.csv')

# O valor mostrado para cada correlação vai de:
# -1 que indica uma correlação negativa (quando se movem em direções opostas)
# +1 uma correlação positiva (quando se movem na mesma direção)

plt.figure(figsize=(5, 5))
sns.heatmap(df.corr(),   #.corr() traz a correlação de Pearson, mostrando um relacionamento linear entre as variáveis
            annot = True,
            fmt = '.2f',
            cmap='Blues')
plt.title('Correlação entre variáveis do dataset de Iris')
plt.show()
````

Sklearn - SelectKBest

```
Retorna um array onde cada elemento dele é uma feature do seu modelo.
Quanto maior o valor, mais importante a feature é para o modelo
```

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

dados = load_iris()
X = dados.data     
y = dados.target 

#Criando conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y) 

# Treinando modelo
model = RandomForestClassifier()
model.fit(X_train, y_train) 

# Verificando em % a importância de cada feature para o modelo
model.feature_importances_

for i in range(len(dados.feature_names)):
   print('Feature:', dados.feature_names[i], model.feature_importances_[i])
````
